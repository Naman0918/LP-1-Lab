```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/Users/namanlabhsetwar/Desktop/Sem 5 Practical/ML- Lab Assignments/LP-1 Dataset/K-Means & Hirarchical Clustering/Hitters.csv')
df.head(5)
```

```python
x=df[['CRuns']]
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_s=scaler.fit_transform(x)
x_s
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters= i, random_state=42, init='k-means++')
    kmeans.fit_predict(x_s)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
```

```python
sns.scatterplot(x='AtBat',y='CRuns',data=df)
plt.xlabel('AtBat')
plt.ylabel('CRuns')
plt.title('hirarchical')
```

```python
cluster=AgglomerativeClustering(n_clusters=3)
df['cluster']=cluster.fit_predict(x_s)
df['cluster']
```

```python
sns.scatterplot(x='AtBat',y='CRuns',hue='cluster',data=df)
plt.xlabel('AtBat')
plt.ylabel('CRuns')
plt.title('hirarchical')
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import numpy as np
```
